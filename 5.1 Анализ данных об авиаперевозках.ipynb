### Шаг 4. Аналитика средствами Python

Извлечем данные из шага 3 в которых содержатся результаты с информацией о модели самолета и количества рейсов для каждой модели самолетов в сентябре 2018 года.

```python
import pandas as pd
```

```python
df_airplane= pd.read_csv("/datasets/query_1.csv")
df_airplane.info()
df_airplane
```

Таблица заполнена полностью, самое большое число рейсов совершено на самолетах Cessna 208 Caravan, меньше всего на Boeing 777-300. Детально изучим информацию после построения графика

Извлечем данные из шага 3 в которых содержатся результаты с информацией о городах и среднем количестве рейсов, прибывающих в город city за день в августе 2018 года.

```python
df_flights = pd.read_csv("/datasets/query_3.csv")
df_flights.info()
df_flights.head()
```

Таблица заполнена полностью. В нашем распоряжении 101 город в аэропортами. Округлим наши даннные в столбце average_flights до целого числа для удобства чтения, присвоим тип int. Детально изучим информацию после построения графика

```python
import numpy as np
```

```python
df_flights['average_flights'] = df_flights['average_flights'].astype('int')
df_flights
```


>**<span style="color:green">Здесь всё структурировано и по делу, класс. </span>**
>


Выберем топ-10 городов по количеству рейсов

```python
df_flights_top = df_flights.sort_values('average_flights', ascending=False).head(10)
df_flights_top
```

В Москве больше всего совершается число рейсов. Это связано со спецификой авиасообщения которое позволяет использовать для перелетов более крупные самолеты (которые на порядок экономичнее в расчете на одного пассажира), чем если бы этот рейс совершался из другого города нашей страны.

Построим график с моделями самолетов и количество рейсов

```python
import pylab as pl
df_airplane_sort = df_airplane.sort_values('flights_amount', ascending=True)
df_airplane_sort.plot(kind='barh', y='flights_amount', x='model', grid=True, figsize=(10, 5))
pl.title('Количество рейсов с вылетом в сентябре 2018 года на каждой модели самолёта')
pl.xlabel("Количество рейсов в сентябре")
pl.ylabel(" ");
```

Большую часть рейсов совершается самолетами малой вместимости Cessna 208 Caravan до 13 пассажиров, Bombardier CRJ-200 до 50 пассажиров, Sukhoi SuperJet-100	до 100 пассажиров.

Построим график зависимости количество рейсов и городов прилета

```python
df_flights_all = df_flights.sort_values('average_flights', ascending=True)
df_flights_all.plot(kind='barh', y='average_flights', x='city', grid=True, figsize=(12, 30))
pl.title('Города по количеству рейсов')
pl.xlabel("Количество рейсов в день")
pl.ylabel(" ");
```

Основная часть рейсов приходится на московские аэропорты и питерский Пулково, чуть меньше загружен аэропорт Новосибирска. Московские аэропорты сильно выделяются на фоне остальных городов, потому что в основном связь больших городов идет через Московскву, как и заграничные перелеты.

Построим график зависимости Топ-10 городов по количеству рейсов прилета

```python
df_flights_top_sort = df_flights_top.sort_values('average_flights', ascending=True)
df_flights_top_sort.plot(kind='barh', y='average_flights', x='city', grid=True, figsize=(10, 5))
pl.title('Топ-10 городов по количеству рейсов')
pl.xlabel("Количество рейсов в день")
pl.ylabel(" ");
```

Разница по числу рейсов Между москвой и вторым по величине городом принимающим самолетыы составляет 4-х кратную величину. Между 3-м городом более чем в 7 раз.
