Определим этаж квартиры со следующими вариантами: первый/последний/другой и добавим их в таблицу

```python
floors_variant = df.loc[:, ['floor', 'floors_total']]

def floor_income(row):
    floor_row = row['floor']
    floors_total_row = row['floors_total']

    if floor_row == 1 or floors_total_row == 1:
            return 'Первый этаж'

    if floor_row > 1 and floors_total_row == floor_row:
        return 'Последний этаж'
             
    return 'Другое'


row_columns_floor = ['floor', 'floors_total']
row = pd.Series(index=row_columns_floor)
df['floors_variant'] = floors_variant.apply(floor_income, axis=1)
#print(df['floors_variant'].value_counts()) 
#print(df['floors_variant'].head(20)) 
#display(df)
```
